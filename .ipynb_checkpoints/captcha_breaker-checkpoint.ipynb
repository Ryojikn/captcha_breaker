```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from PIL import Image
import numpy as np

ticker = "ITSA4"
ticker = ticker.lower()
balance_url = f"https://www.fundamentus.com.br/balancos.php?papel={ticker}&tipo=1"

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(balance_url)
```

```python
print(driver.title)
```

```python
from selenium.webdriver.common.by import By


driver.save_screenshot('screenshot.png')
captcha_object = driver.find_element(by=By.CSS_SELECTOR, value="img.captcha")
captcha_location = captcha_object.location

print(captcha_location)
```

```python
import cv2
screenshot_image = cv2.imread('screenshot.png')
## changing from bgr to grayscale
screenshot_image = cv2.cvtColor(screenshot_image, cv2.COLOR_BGR2GRAY)
```

```python
## for mac m1 screen 13 inch, it seems that multiplying by 2 will give you the exact
# position for the desired crop, but in another situation you might test another configurations

y = captcha_location["y"]*2
x = captcha_location["x"]*2

## the ideal size of the captcha considering your monitor
## use mspaint or another tool to checkup based on screenshot.png
h = 130
w = 400

## cropping the image
cropped_img = screenshot_image[y:y+h, x:x+w]
```

```python
Image.fromarray(cropped_img)
```

```python
import pytesseract
```

```python
cropped_img.mean()
```

```python
pytesseract.image_to_string(cropped_img).strip().replace(' ','')
```

## OK, this was awful, let's do some processing

```python
##saving actual image
Image.fromarray(cropped_img).save('cropped_img.png')
```

```python
img = Image.open('cropped_img.png').convert('LA')
```

Here we convert to grayscale, as when considering grayscale we only have luminance.

Zero luminance stands for black color, and maximum luminance stands for white color.

For a higher detail, please check [here](https://medium.com/@overclockedBrains/dealing-with-captchas-b2d5af526566).

```python
## change this
# 
threshold = 100
```

```python
import cv2

data = np.array(img)
data[np.where( data > threshold )]=255                   # Line 9 
Image.fromarray(data).save('cropped_img.png')
new_img = cv2.imread('cropped_img.png', 0)
```

```python
Image.fromarray(new_img)
```

```python
pytesseract.image_to_string(new_img).strip().replace(' ','')
```

### Tesseract still isn't getting it.

```python
import cv2


inverted_img_array = cv2.threshold(new_img, threshold,255, cv2.THRESH_BINARY + cv2.THRESH_BINARY_INV)[1]
```

```python
Image.fromarray(inverted_img_array)
```

```python
pytesseract.image_to_string(inverted_img_array).strip().replace(' ','')
```

```python
kernel = np.ones((5,5), dtype=np.uint8)
new_img_dilated = cv2.dilate(inverted_img_array, kernel, iterations=1)
```

```python
Image.fromarray(new_img_dilated).save('edited_img.png')
Image.fromarray(new_img_dilated)
```

```python
pytesseract.image_to_string(new_img_dilated).strip().replace(' ','')
```

```python
import easyocr


reader = easyocr.Reader(lang_list=['pt'], gpu=False)
```

```python
reader.readtext('edited_img.png')[0][1]
```

```python
# Trying mexican hat filtering

# Creating maxican hat filter
mexican_hat_filter = np.array([[0,0,-1,0,0],[0,-1,-2,-1,0],[-1,-2,16,-2,-1],[0,-1,-2,-1,0],[0,0,-1,0,0]])

# Applying cv2.filter2D function on our image
mexican_hat_img1=cv2.filter2D(reinverted_img_array,-1,mexican_hat_filter)
Image.fromarray(mexican_hat_img1)
```

```python
Image.fromarray(mexican_hat_img1).save('edited_img.png')
```
